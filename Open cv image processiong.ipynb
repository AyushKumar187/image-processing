```python
import numpy as np                           
from skimage.io import imread                
from skimage.color import rgb2lab, lab2rgb   
import matplotlib.pyplot as plt             
Input Image

im = imread('cat.png')
plt.title('Original Image')
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.show()im.shape
(512, 512, 3)
(height, widht, colorChannels)

im[:,:,0]: Read
im[:,:,1]: Green
im[:,:,2]: Blue
Convert the Image Color Space: skimage.color.rgb2lab

Conversion from RGB color space into CIELAB color space.

# RGB to Gray Scale
imLab = rgb2lab(im)
imLab.shape
(512, 512, 3)
im[:,:,0]: Intensity (L)
im[:,:,1]: Green-red (A)
im[:,:,2]: Blue-yellow (B)
Gray Scale

#  Gray Scale
imGray = imLab.copy()
imGray[:,:,1] = 0   
imGray[:,:,2] = 0   
imGray = lab2rgb(imGray)
plt.title('Gray Scale Image')
plt.imshow(imGray)
plt.axis('off')
plt.show()# Brighter
imBrighter = imLab.copy()

imDarker = imLab.copy()
imBrighter[:,:,0] = imBrighter[:,:,0] + 50
imBrighter = lab2rgb(imBrighter)
plt.title('Brighter Image')
plt.imshow(imBrighter)
plt.axis('off')
plt.show()
imDarker[:,:,0] = imDarker[:,:,0] - 50
imDarker = lab2rgb(imDarker)
plt.title('Darker Image')
plt.imshow(imDarker)
plt.axis('off')
plt.show()
imInvert = imLab.copy()
imInvert[:,:,0] = np.max(imInvert[:,:,0]) - imInvert[:,:,0]
imInvert = lab2rgb(imInvert)
plt.title('Inverted Image')
plt.imshow(imInvert)
plt.axis('off')
plt.show()
```
